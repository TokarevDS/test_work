# Задание 1

## Описание

<br>
У нас есть сервис поиска дешевых авиабилетов. Основная воронка: юзер зашел на сайт, нажал "искать билеты", открыл конкретный билет, выбрал предложение конкретного партнера и перешел к нему на сайт, купил билет на сайте. Партнер платит нам комиссию за покупку билета.
<br><br>
На экране билета (скрин того, о чём речь, ниже) в десктопной версии сайта предложения от разных партнеров расположены от меньшей цены к большей.
<br>
Как иначе можно расположить предложения от разных партнеров на один и тот же билет? Как оценить эффективность предложенного подхода?

# Задание 2

## Описание

Мы хотим провести A/B-тест с двумя группами - контрольной и тестовой.
<br>
Есть ключевая метрика СTR (clicks / views) и исторические данные по ней:
<br>
- Файл <i>task_2_history_control.csv</i> - данные по периоду, когда не было A/B-тестов, можно считать, что метрика очень стабильна.
- Файл <i>task_2_history_old_ab.csv</i> - показатели по тестовой группе предыдущего A/B-теста, где CTR упал на 15%.

В A/B-тесте хотим статистически значимо зафиксировать увеличение СTR на 10%. Допустимая ошибка первого рода - 5%, второго рода - 20%.
<br><br>
1. Рассчитайте объем выборки, необходимый в каждой из групп, если в контрольной группе будет 90% юзеров, а в тестовой 10%.
<br><br>
2. Мы провели тест и получили результаты (файл <i>task_2_results.csv</i>, одна строка - один юзер, дубликатов по юзерам нет). Рассчитайте p-value, обоснуйте выбор критерия.
